* The .get_dummies() function in Pandas is used to convert categorical data (such as strings or integers) into numerical data by creating "dummy" or "one-hot encoded" variables. This is especially useful when preparing data for machine learning models that require numerical input.

```python
df.shape
```

```python
df['C1'].unique()
```

```python
df['C1'].unique().size
```

```python
#Converted the values in int data type, default was True/False. Now it is giving 1 if present, 0 otherwise.
df=pd.get_dummies(df, columns=['C1'], dtype='uint8')
```
