Bokeh is a Python library used for creating interactive, web-based visualizations that are highly customizable and scalable. It allows users to generate complex visualizations, such as plots, graphs, dashboards, and data applications, directly in a web browser. Bokeh is designed to work well with large and streaming datasets, making it ideal for real-time data visualization.

```python
from bokeh.plotting import output_notebook, show, figure
```

```python
output_notebook()
p=figure(title='sample', width=600, height=400)
p.line([1,2,3],[4,5,6], line_width=3)
show(p)
```

* If we use output_file('name.html') instead of output_notebook(), it will give plot at webpage named name.html. Firstly import output_file as well.

```python
p=p=figure(title='circle', width=600, height=400)
p.circle(1,4,size=5,color='red')
show(p)
```

```python
import numpy as np
```

```python
p=figure(title='Graphs',width=600,height=400, x_axis_label='X',y_axis_label='Y')
x=np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
p.triangle(x,y1,size=5,color='red', legend_label='sin(x)')
p.circle(x,y2,size=5,color='green', legend_label='cos(x)')
show(p)
```

#gridplot<br>
In Bokeh, .gridplot is used to create a grid of subplots (multiple plots arranged in a grid format). It is especially helpful when you want to display multiple plots in a structured layout. This allows you to organize several plots in rows and columns within a single view, making it easier to compare multiple visualizations.

```python
 from bokeh.layouts import gridplot
```

```python
x=np.linspace(1,10,50)
y1=np.sin(x)
y2=np.cos(x)
g1=figure(title='Graph1',width=600,height=400, x_axis_label='X',y_axis_label='Y')
g1.triangle(x,y1,size=5,color='red', legend_label='sin(x)', alpha=0.5)

g2=figure(title='Graph2',width=600,height=400, x_axis_label='X',y_axis_label='Y')
g2.square(x,y2,size=5,color='green', legend_label='cos(x)', alpha=0.5)

p=gridplot([[g1,g2]])
show(p)
```
