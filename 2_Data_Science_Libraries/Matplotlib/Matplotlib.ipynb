#Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

* plt.style is a module that provides a way to apply predefined or custom styles to your plots, making it easy to change the appearance of your plots with minimal effort.



```python
plt.style.available
```

```python
x=np.linspace(0,10,1000)
y=np.sin(x)
plt.style.use('classic')
plt.plot(x,y)
```

```python
plt.style.use('dark_background')
plt.plot(x,y)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
plt.plot(x,y)
```

* The plt.figure() function is used to create a new figure (a blank canvas) on which you can draw plots. It allows you to specify various properties of the figure, such as its size, resolution, and background color.

* The plt.axis() function is used to set or get properties of the axes of the current plot, such as the limits, scaling, and visibility of the axes.


```python
fig=plt.figure()
```

```python
ax=plt.axes()
```

#plt.plot() vs ax.plot()<br>
* plt.plot() is part of the pyplot interface, which is a state-based interface in Matplotlib. It automatically manages figures and axes for you, making it easier to create plots without worrying about the underlying structure. This is often referred to as the "pyplot" style or the "MATLAB-like" style. <br>

* ax.plot() is a way to draw a line plot on a specific "part" of a figure in Matplotlib.This is often preferred for more complex or multi-plot figures, as it allows for more precise control. <br>
<b>If we dont specify any axes as such, both works the same.</b>



```python
fig=plt.figure(figsize=(10,4))  #figsize=(width,height)
ax=plt.axes()
x=np.linspace(0,10,1000)
y=np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('classic')
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.plot(x,np.cos(x-0.5))
```

```python
plt.plot(np.tan(x))
```

#Types to specify colours<br>
1) color='red'<br>
2) color='r'<br>
3) color=(1.0,0,0) {in form of RGB tuple}

```python
plt.plot(x,y,color='red')
plt.plot(x,np.cos(x),color='b')
plt.plot(x,np.cos(x-0.5),color=(0,1,0))
```

#Linestyle

```python
plt.plot(x,y,linestyle='dashed')
```

```python
plt.plot(x,y,'-g')
```

#Linewidth<br>
Increases thickness of specified line

```python
plt.plot(x,y)
plt.plot(x,np.cos(x),linewidth=5)
```
