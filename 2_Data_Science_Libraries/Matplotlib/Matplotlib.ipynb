#plt.plot() vs ax.plot()<br>
* plt.plot() is part of the pyplot interface, which is a state-based interface in Matplotlib. It automatically manages figures and axes for you, making it easier to create plots without worrying about the underlying structure. This is often referred to as the "pyplot" style or the "MATLAB-like" style. <br>

* ax.plot() is a way to draw a line plot on a specific "part" of a figure in Matplotlib.This is often preferred for more complex or multi-plot figures, as it allows for more precise control. <br>
<b>If we dont specify any axes as such, both works the same.</b>



```python
fig=plt.figure(figsize=(10,4))  #figsize=(width,height)
ax=plt.axes()
x=np.linspace(0,10,1000)
y=np.sin(x)
ax.plot(x,y)
```

```python
plt.style.use('classic')
plt.plot(x,y)
plt.plot(x,np.cos(x))
plt.plot(x,np.cos(x-0.5))
```

```python
plt.plot(np.tan(x))
```

* DPI:<br>
DPI stands for "Dots Per Inch" and is a measure of the resolution of a figure. It determines how many pixels (dots) are contained within an inch on your display or in a saved figure. The higher the DPI, the more detail you will have in the figure, but it will also result in a larger file size.



```python
fig=plt.figure(figsize=(5,3),dpi=100)
ax=plt.axes()
x=np.linspace(0,10,1000)
y=np.sin(x)
ax.plot(x,y)
```

```python
fig=plt.figure(figsize=(5,3),dpi=200)
ax=plt.axes()
ax.plot(x,y)
```

#xlim and ylim <br>
These functions are used to set or get the limits of the x-axis and y-axis, respectively.<br>

plt.xlim([xmin, xmax]) <br>
plt.ylim([ymin,ymax])

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.xlim(0,8)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.ylim(0,2)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.xlim(0,8)
plt.ylim(-2,2)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,x+3)
plt.xlim(0,8)
```

* Reversing the limits reverse the plot

```python
plt.figure(figsize=(5,3))
plt.plot(x,x+3)
plt.xlim(8,0)
```

#.axis()<br>
The .axis() function is a more general tool that can set both x and y limits at once and control other aspects of the plot axes. <br>

Purpose: Sets the limits for both axes, aspect ratio, and more.<br>
plt.axis([xmin, xmax, ymin, ymax])

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.axis([0,8,-2,2])
```

* plt.axis('tight'):<br>
The plt.axis('tight') command in Matplotlib adjusts the axis limits to fit the data tightly. This means that Matplotlib automatically sets the axis limits so that all data points are visible, with no extra space around the edges of the plot.

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.axis('tight')
```

* The plt.axis('equal'):<br>
This command in Matplotlib ensures that the scaling of the x and y axes is equal, meaning that one unit on the x-axis is the same length as one unit on the y-axis. This results in plots where circles look like circles (instead of ellipses) and squares look like squares (instead of rectangles), which is important for accurately representing data with geometric shapes.

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
plt.axis('equal')
```

#Labels and Legends<br>
In Matplotlib, labels and legend are used to add clarity to plots by labeling data series and creating a legend that describes the plotted data. <br>
<b>Without using plt.legend(), we can't see effect of label</b>

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label='sin(x)')
plt.legend(loc='lower right')
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend(loc='right')
```

* xlabel() and ylabel()<br>
To give labels to x-axis and y-axis

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
```

plt.title()<br>
To give title to whole plots.

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,label='sin(x)')
plt.plot(x,np.cos(x),label='cos(x)')
plt.legend()
plt.title('sin(x) and cos(x) graphs')
```

#Pyplot vs axes<br>
plt.xlabel --> ax.set_xlabel()<br>
plt.ylabel --> ax.set_ylabel()<br>
plt.xlim --> ax.set_xlim()<br>
plt.ylim --> ax.set_ylim()<br>
plt.title --> ax.set_title()<br><br>
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;OR <br><br>
ax.set(xlabel='', ylabel='', xlim=(), ylim=(), title='')

```python
fig=plt.figure(figsize=(5,3))
ax=plt.axes()
ax.set_xlabel('x')
ax.set_ylabel('sin(x)')
ax.set_title('sin(x) graph')
ax.set_xlim(0,10)
ax.set_ylim(-1,1)
ax.plot(x,y)
```

#Scatter<br>
The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis

```python
x=np.array([1,2,3,4])
y=np.array([4,2,6,4])
plt.figure(figsize=(5,3))
plt.scatter(x,y)
```

```python
plt.figure(figsize=(5,3))
x=np.array([1,2,3,4])
y=np.array([4,2,6,4])
plt.scatter(x,y)
x=np.array([1,2,3,4])
y=np.array([2,4,5,8])
plt.scatter(x,y)
```

* colormap<br>
A colormap in Matplotlib is a range of colors used to map data values to colors in visualizations like heatmaps, scatter plots, or surface plots.

```python
plt.colormaps()
```

```python
plt.figure(figsize=(5,3))
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
plt.scatter(x,y,c=colors,cmap='viridis')
```

* s<br>
You can change the size of the dots with the s argument.



```python
plt.figure(figsize=(5,3))
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
size=np.random.rand(100)*100
plt.scatter(x,y,c=colors,cmap='viridis',s=size)
```

* Alpha<br>
You can adjust the transparency of the dots with the alpha argument.



```python
plt.figure(figsize=(5,3))
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
size=np.random.rand(100)*100
plt.scatter(x,y,c=colors,cmap='viridis',s=size,alpha=1)
```

```python
plt.figure(figsize=(5,3))
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
size=np.random.rand(100)*100
plt.scatter(x,y,c=colors,cmap='viridis',s=size,alpha=0.2)
```

* Colorbar<br>
A colorbar in Matplotlib is a color scale that provides a reference for the mapping of data values to colors in a plot.

```python
plt.figure(figsize=(5,3))
x=np.random.randn(100)
y=np.random.randn(100)
colors=np.random.rand(100)
size=np.random.rand(100)*100
plt.scatter(x,y,c=colors,cmap='viridis',s=size,alpha=0.8)
plt.colorbar()
```

```python
plt.figure(figsize=(5,3))
x=np.random.randn(1000)
y=np.random.randn(1000)
colors=np.random.rand(1000)
size=np.random.rand(1000)*100
plt.scatter(x,y,c=colors,cmap='autumn',s=size,alpha=0.9)
plt.colorbar()
```
