* np.contour(): plt.contour is a function in Matplotlib used to create contour plots, which are a way of representing three-dimensional data in two dimensions. A contour plot is a graphical technique to represent a 3D surface by plotting constant z slices, called contours, on a 2D plane. These contours represent different levels of the z-axis values and are drawn as lines.<br><br>
<b>Syntax: plt.contour(X, Y, Z, levels=None, cmap=None)</b><br>
X, Y: These are 2D arrays (typically created using np.meshgrid) that represent the grid of points over which the function is evaluated. <br>
Z: A 2D array of the same shape as X and Y, containing the function values at each point in the grid.<br>
levels: Specifies the number of contour lines to draw. If not specified, Matplotlib will choose levels automatically. <br>


```python
#No specified level
plt.figure(figsize=(5,3))
X,Y=np.meshgrid(np.linspace(0,5,10),np.linspace(0,5,10))
Z=np.sin(X)+np.cos(Y)
plt.contour(X,Y,Z)
```

```python
#20 levels
plt.figure(figsize=(5,3))
X,Y=np.meshgrid(np.linspace(0,5,10),np.linspace(0,5,10))
Z=np.sin(X)+np.cos(Y)
plt.contour(X,Y,Z,20)
```

* 3D Spiral<br>
1) theta = np.linspace(-4 * np.pi, 4 * np.pi, 100): Creates an array of 100 values for theta, ranging from -4pi to 4pi This array represents the angle parameter for the spiral.<br>
2) z = np.linspace(-2, 2, 100): Creates an array of 100 values for z, ranging from -2 to 2. This array represents the height or vertical position of the spiral.<br>
3) r = z**2 + 1: Computes the radius of the spiral at each z value. The radius is a function of z and increases quadratically. This formula determines how far the spiral extends from the origin in the x-y plane as z changes.<br>
4) x = r * np.sin(theta): Calculates the x-coordinates of the spiral. It uses the radius r and the angle theta to compute the x-values using the sine function.<br>
5) y = r * np.cos(theta): Calculates the y-coordinates of the spiral. It uses the radius r and the angle theta to compute the y-values using the cosine function.

```python
fig = plt.figure(figsize=(5, 3))
ax = fig.add_subplot(projection='3d')

theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

ax.plot(x, y, z)
```
