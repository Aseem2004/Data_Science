Folium is a Python library used for visualizing data on interactive maps by leveraging Leaflet.js (a JavaScript library for maps). It allows users to create beautiful maps with markers, popups, paths, and other geographical features, making it particularly useful for geospatial data visualizations.

```python
import folium
```

```python
map=folium.Map(width=400, height=300)
map
```

* Pointing to specific coordinates

```python
map=folium.Map(location=[16.7850,80.8488], width=400, height=300)
map
```

* Adding marker
 * popup: This creates a popup box with text that appears when you click on the marker.

```python
map=folium.Map(location=[37.7749, -122.4194], zoom_start=12, height=300, width=400)
folium.Marker([37.7749, -122.4194], popup="Location").add_to(map)
map
```

* Adding a circular marker

```python
map=folium.Map(location=[37.7749,-122.4194], zoom_start=12, height=300, width=400)

folium.CircleMarker(
    location=[37.7749,-122.4194],
    radius=50,
    popup="San Francisco",
    color='blue',
    fill_color='red'
).add_to(map)

map
```

#Analyzing Covid 19 Dataset on Globe

```python
import pandas as pd
```

```python
df=pd.read_csv('covid_19_till_19_May_2020.csv')
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df2=df[df['Date']==max(df['Date'])]
```

```python
df2
```

```python
map=folium.Map(height=400, width=600)

for i in range(len(df2)):
  folium.CircleMarker(
      location=[df2.iloc[i]['Lat'], df2.iloc[i]['Long']],
      radius=int(df2.iloc[i]['Confirmed']**0.2),
      popup=df2.iloc[i]['Country/Region'], color='crimson', fill_color='crimson').add_to(map)

map
```

```python
map=folium.Map(height=400, width=600)

for i in range(len(df2)):
  folium.CircleMarker(
      location=[df2.iloc[i]['Lat'], df2.iloc[i]['Long']],
      radius=int(df2.iloc[i]['Deaths']**0.2),
      popup=df2.iloc[i]['Country/Region'], color='crimson', fill_color='crimson').add_to(map)

map
```

```python
map=folium.Map(height=400, width=600)

for i in range(len(df2)):
  folium.CircleMarker(
      location=[df2.iloc[i]['Lat'], df2.iloc[i]['Long']],
      radius=int(df2.iloc[i]['Confirmed']**0.2),
      popup=df2.iloc[i]['Country/Region'], color='crimson', fill_color='crimson').add_to(map)

map
```
