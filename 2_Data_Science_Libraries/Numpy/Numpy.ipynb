#plt.imread(): <br>
plt.imread is a function from the matplotlib.pyplot module in Python. It is used to read (load) an image file into a NumPy array. This function is commonly used for image processing tasks, data visualization, and loading image data for analysis. <br>
* image: A NumPy array containing the image data. The shape of the array depends on the image type: <br>
1) For RGB images, the array will have the shape (height, width, 3). <br>
2) For RGBA images, the array will have the shape (height, width, 4).


#RGB vs RGBA <br>

### RGB (Red, Green, Blue)

- **Components**: RGB stands for Red, Green, and Blue, which are the three primary colors used in this model. Each pixel in an RGB image is represented by a combination of these three color channels. <br>
- **Channels**: An RGB image has three channels: <br>
  - **Red**: Controls the intensity of the red component. <br>
  - **Green**: Controls the intensity of the green component. <br>
  - **Blue**: Controls the intensity of the blue component.<br>

### RGBA (Red, Green, Blue, Alpha) <br>

- **Components**: RGBA is similar to RGB but includes an additional **Alpha** channel that represents the opacity or transparency of each pixel. <br>
- **Channels**: An RGBA image has four channels: <br>
  - **Red**: Controls the intensity of the red component. <br>
  - **Green**: Controls the intensity of the green component. <br>
  - **Blue**: Controls the intensity of the blue component. <br>
  - **Alpha**: Controls the transparency level of the pixel. An alpha value of 0 means fully transparent, while 255 (or 1.0 for float values) means fully opaque.<br>

````python
fig, ax = plt.subplots(1,3)
img2 = img.copy()
img2[:,:,2] = img[:,:,2]*10
ch = ['Red','Green','Blue']
for i,axi in enumerate(ax.flat):
    axi.imshow(img2[:,:,i],plt.cm.gray) #Notice colour map hereis gray which gives the final output a grayish look otherwise it would have been same as above
    axi.set_xlabel(ch[i],color=ch[i].lower())
    axi.set(xticks=[], yticks=[])
    fig.set_size_inches(20,50)
````

This code is intended to display the individual color channels (Red, Green, and Blue) of an image in grayscale using Matplotlib. The code modifies the blue channel by multiplying it by 10, which increases its intensity, and then it plots the red, green, and modified blue channels separately.

### Step-by-Step Explanation

1. **Creating Subplots:**
   ```python
   fig, ax = plt.subplots(1, 3)
   ```
   This line creates a figure with 1 row and 3 columns of subplots (`ax` is a list of Axes objects).

2. **Copying and Modifying the Image:**
   ```python
   img2 = img.copy()
   img2[:, :, 2] = img[:, :, 2] * 10
   ```
   A copy of `img` is made to `img2`, and then the blue channel (third channel, index `2`) of `img2` is multiplied by 10, increasing its intensity. This modification affects only `img2`, leaving the original `img` unchanged.

3. **Plotting the Channels:**
   ```python
   ch = ['Red', 'Green', 'Blue']
   for i, axi in enumerate(ax.flat):
       axi.imshow(img2[:, :, i], plt.cm.gray)
       axi.set(xticks=[], yticks=[])
       axi.set_xlabel(ch[i], color=ch[i].lower())
   fig.set_size_inches(20, 50)
   ```
   - **`ch = ['Red', 'Green', 'Blue']`**: This list is used to label each subplot with the corresponding channel name.
   - **Loop through `ax.flat`:** `ax.flat` allows you to iterate over all subplots. The loop runs 3 times, once for each color channel.
     - **`axi.imshow(img2[:, :, i], plt.cm.gray)`**: Displays each channel of `img2` in grayscale (`plt.cm.gray`) in its respective subplot. `img2[:, :, i]` selects the `i`-th channel (Red, Green, or Blue).
     - **`axi.set(xticks=[], yticks=[])`**: Removes the axis ticks for a cleaner appearance.
     - **`axi.set_xlabel(ch[i], color=ch[i].lower())`**: Labels each subplot with the name of the channel (Red, Green, or Blue) in the corresponding color.
   - **`fig.set_size_inches(20, 50)`**: Sets the figure size to 20 inches wide by 50 inches tall, which is quite large and may not be necessary depending on your display. This line controls the overall size of the figure, but you might want to adjust these values based on your needs.
