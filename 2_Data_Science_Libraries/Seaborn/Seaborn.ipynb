```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#kdeplot <br>
sns.kdeplot() is a function in the Seaborn library used to generate Kernel Density Estimate (KDE) plots. A KDE plot visualizes the probability density function (PDF) of a continuous variable, which shows how the data is distributed.<br>
sns.kdeplot(data=None, x=None, y=None, shade=False)<br>
* data: The dataset from which variables will be plotted.<br>
* x: The data variable to be plotted on the x-axis.<br>
* y: (Optional) The second variable for a bivariate KDE plot.<br>
* shade: (deprecated, use fill) If True, it fills the area under the KDE curve.



```python
df=pd.DataFrame({'X':np.random.normal(2,3,1000),
                'Y':np.random.normal(1,2,1000)})
```

```python
df.head()
```

```python
plt.figure(figsize=(5,3))
sns.set()
for col in df:
  sns.kdeplot(df[col],fill=True)
```

#distplot<br>
sns.distplot() was a function in Seaborn used to plot both a histogram and a Kernel Density Estimate (KDE) plot on the same graph. However, as of Seaborn version 0.11.0, sns.distplot() is deprecated and has been replaced by two newer functions:<br>

sns.histplot(): For histograms.<br>
sns.kdeplot(): For KDE plots.

```python
plt.figure(figsize=(5,3))
for col in df:
  sns.distplot(df[col])
```

#jointplot<br>
A jointplot in Seaborn is like combining two types of plots into one to show the relationship between two variables and how each one is distributed. <br>

* Main Plot (middle part): It shows how two things are related, like a scatter plot (which shows dots for each pair of values).<br>
* Marginal Plots (side parts): It also adds small plots on the top and the right, showing how each variable is spread out individually (like a histogram or smooth line called a KDE).

```python
plt.figure(figsize=(5,3))
sns.jointplot(x='X',y='Y',data=df,kind='kde')
```

#pairplot<br>
A pairplot is a grid of scatter plots (or other plots) that shows the relationships between all pairs of variables in a dataset. It also adds diagonal plots showing the distribution of each variable.



```python
iris=sns.load_dataset('iris')
sns.pairplot(iris,hue='species')

#Hue: The hue='species' argument colors the scatter plots according to the flower species, so you can easily see how the species are different based on the measurements.
```

```python
iris.head()
```

#relplot<br>
A relplot (short for "relationship plot") in Seaborn is a versatile function for creating different types of plots that show the relationship between two or more variables. Its built to handle both scatter plots and line plots, depending on the type of data you're working with.<br><br>
Syntax: sns.relplot(x='x_var', y='y_var', data=data, kind='',ci='')<br>
* kind is either 'scatter'(def) or 'line'<br>
* The ci parameter stands for confidence interval. It is used to display the uncertainty around the estimated line in a line plot (or scatter plot with trend lines).

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,style='species')
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='sepal_length')
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='sepal_length',style='species')
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,kind='line')
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',kind='line',ci=None)
```

```python
plt.figure(figsize=(5,3))
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species',kind='line',ci=None, style='species',markers=True)
```

* col <br>
The col parameter in Seaborn's relplot() is used to create multiple plots (facets) based on the unique values of a categorical variable. It allows you to split your data into different subplots (columns), where each subplot corresponds to one category of the variable.

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#total vs tip according to smoker during both lunch and dinner time
plt.figure(figsize=(5,3))
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='time')
```

* .query() allows you to filter a DataFrame using a query string that resembles SQL syntax.

```python
tips['day'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips.query("day=='Sun'"),hue='smoker',col='time')
```

#catplot<br>
sns.catplot is a versatile function in Seaborn used to create categorical plots. It can generate several types of plots (like bar plots, box plots, violin plots, etc.) to visualize the distribution of categorical data and the relationships between categorical and numerical variables.<br>
Syntax: sns.catplot(x='', y='', data=data, kind='')<br>
* kind:<br>
 * 'strip': Strip plot (like a scatter plot for categorical data).<br>
 * 'swarm': Swarm plot (a variation of a strip plot that avoids overlapping).<br>
 *'box': Box plot.<br>
 *'violin': Violin plot.<br>
 *'bar': Bar plot.<br>
 *'count': Count plot (to show the counts of observations in each categorical bin).


```python
sns.catplot(x='day',y='total_bill',data=tips,kind='strip')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='swarm',hue='sex')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin',hue='sex')
```

#heatmap<br>
sns.heatmap is a function in Seaborn that creates a heatmap, which is a graphical representation of data where individual values are represented by colors. Heatmaps are particularly useful for visualizing matrix-like data, allowing you to see patterns, correlations, and relationships at a glance.<br>
Syntax: sns.heatmap(data, annot=False) <br>
* annot: sns.heatmap(data, annot=False, fmt='.2f', cmap='coolwarm', linewidths=0.5)



```python
df=pd.DataFrame(np.random.rand(10,10), columns=list('ABCDEFGHIJ'))
```

```python
df
```

```python
sns.heatmap(df,annot=True,annot_kws={'size':10}) #To determine size of text
```
