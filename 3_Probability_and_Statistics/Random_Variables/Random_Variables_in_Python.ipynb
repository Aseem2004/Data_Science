#Bernoulli Trials

```python
import numpy as np
```

```python
def Ber(p):
  x=np.random.rand()
  if x<=p:                #If success, return 1
    return 1
  else:                   #If failure, return 0
    return 0
```

```python
for i in range(10):
  print(Ber(0.7),end=' ')   #p: probability of success (can be thought as threshold value in logistic regression, below 0.7 all are 1 and above are 0, hence 70% chance of getting 1)
```

* Calculating % of chances of getting 1 based on increased number of simulations

```python
ones=0
zeroes=0
p=0.7
n=10

for i in range(n):
  x=int(Ber(p))
  if x==1:
    ones+=1
  else:
    zeroes+=1

print((ones/n)*100)
```

* As we increase trials, % gets closer to n

```python
ones=0
zeroes=0
p=0.7
n=100000

for i in range(n):
  x=int(Ber(p))
  if x==1:
    ones+=1
  else:
    zeroes+=1

print((ones/n)*100)
```

#Geometric Trials

```python
def GT(p):
  x=1
  while(1):
    if(Ber(p)==1):
      break
    else:
      x+=1

  return x
```

```python
for i in range(10):
  print(GT(0.5), end=' ')
```

```python
n=1000
p=0.7             #Probability of success = 0.7
l=[]

for i in range(n):
  l.append(GT(p))

print(l)
```

```python
print(np.mean(l))
```

#Binomial Random Variable

```python
def Binomial(n,p):
  x=(np.random.rand(n)<=p).sum()
  return x
```

```python
Binomial(100,0.5)   #Out of 100 trials, chances of getting success is in 53, if chances of success = 0.5 (50%)
```

```python
N=10000
n=100
p=0.7
Bin=np.zeros(N)

for i in range(N):
  Bin[i]=Binomial(n,p)

print(Bin)          #There were N testcases performed. In each one of them, n trials were performed and chances of getting success is p
```

```python
np.mean(Bin)
```

```python
plt.hist(Bin,bins=10)     #Roughly shows that in more than 3000 values (out of N=10000), atleast there were 70 success (out of n=100)
```

```python
sns.kdeplot(Bin, fill=True)
```

#Exponential Distribution

```python
x=np.linspace(0,3,1000)
lmda=0.5
y=lmda*np.exp(-lmda*x)
plt.plot(x,y)
```

```python
x=np.linspace(0,3,1000)
lmda1=0.1
lmda2=0.5
lmda3=0.8
lmda4=1
lmda5=2

y1=lmda1*np.exp(-lmda1*x)
y2=lmda2*np.exp(-lmda2*x)
y3=lmda3*np.exp(-lmda3*x)
y4=lmda4*np.exp(-lmda4*x)
y5=lmda5*np.exp(-lmda5*x)

plt.plot(x,y1, label=lmda1)
plt.plot(x,y2, label=lmda2)
plt.plot(x,y3, label=lmda3)
plt.plot(x,y4, label=lmda4)
plt.plot(x,y5, label=lmda5)

plt.legend()      #Larger the lambda, larger the peak but larger decay as well to maintain area under curve as 1
```

* Exponential Distribution as builtin function in Numpy<br>
np.random.exponential(scale, size): Here parameter you provide is not directly , but rather the scale parameter, which is the inverse of  (scale=1/). And size describes shape of the output array. For example, size=10 will generate 10 random values.


```python
np.random.exponential(scale=0.5, size=3)
```

```python
x=np.random.exponential(scale=2, size=10000)   #Here scale=2 means lmda=0.5
plt.hist(x, bins=100, density=True)
```

#Gaussian/Normal Distribution

```python
x=np.linspace(-10,10,1000)
mu=0
sgma=0.5

y=np.exp(-(x-mu)**2/(2*sgma**2))/(sgma*np.sqrt(2*np.pi))
plt.plot(x,y)
```

```python
#Changing values of mu, it represents where data would be centered

x=np.linspace(-10,10,1000)
muz=np.arange(-4,5,2)
sgma=0.5

for mu in muz:
  y=np.exp(-(x-mu)**2/(2*sgma**2))/(sgma*np.sqrt(2*np.pi))
  plt.plot(x,y,label=mu)
plt.legend()
```

```python
#Changing values of sgma, it represents rate of decay of graph

x=np.linspace(-10,10,1000)
mu=0
sgmaz=np.arange(1,10,2)

for sgma in sgmaz:
  y=np.exp(-(x-mu)**2/(2*sgma**2))/(sgma*np.sqrt(2*np.pi))
  plt.plot(x,y,label=sgma)
plt.legend()
```

* Gaussian Distribution as builtin function in Numpy<br>
np.random.normal(loc,scale,size): where loc:mu, scale:sgma, size:number of points

```python
np.random.normal(0,1,5)
```

```python
x=np.random.normal(0,1,1000)    #mu=0 and sgma=1
plt.hist(x, density=True, bins=50)
```
